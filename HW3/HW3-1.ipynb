```python
__author__ = 'arp504'

import pylab as pl
import numpy as np
%pylab inline
```

```python
mysizes=np.random.random_integers(10,2000,100)
mysizes.sort()
mymean = 100
print "These are the 100 random sizes obtained from 10 and 2000"
print (mysizes)
```

```python
allmeans = []
distributions={'normal':{},'poisson':{},'binomial':{},'chisq':{},'gamma':{},'exponential':{}}
for n in mysizes:
    distributions['normal'][n] = np.random.normal(mymean,10,n)
    distributions['poisson'][n] = np.random.poisson(mymean,n)
    distributions['binomial'][n] = np.random.binomial(2*mymean,0.5,n)
    distributions['chisq'][n] = np.random.chisquare(mymean,n)
    distributions['gamma'][n] = np.random.gamma(mymean,1,n)
    distributions['exponential'][n] = np.random.exponential(mymean,n)
for n in mysizes:
    allmeans.append(distributions['normal'][n].mean())
    allmeans.append(distributions['poisson'][n].mean())
    allmeans.append(distributions['binomial'][n].mean())
    allmeans.append(distributions['chisq'][n].mean())
    allmeans.append(distributions['gamma'][n].mean())
    allmeans.append(distributions['exponential'][n].mean())   
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axnormal_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['normal']:
    axnormal_mu_n.plot(nn,distributions['normal'][nn].mean(),'o')
    axnormal_mu_n.set_xlabel('sample size')
    axnormal_mu_n.set_ylabel('sample mean')
    axnormal_mu_n.set_title('Means for Normal Distributions')
    axnormal_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axbinomial_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['binomial']:
    axbinomial_mu_n.plot(nn,distributions['binomial'][nn].mean(),'o')
    axbinomial_mu_n.set_xlabel('sample size')
    axbinomial_mu_n.set_ylabel('sample mean')
    axbinomial_mu_n.set_title('Means for Binomial Distributions')
    axbinomial_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axpoisson_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['poisson']:
    axpoisson_mu_n.plot(nn,distributions['poisson'][nn].mean(),'o')
    axpoisson_mu_n.set_xlabel('sample size')
    axpoisson_mu_n.set_ylabel('sample mean')
    axpoisson_mu_n.set_title('Means for Poisson Distributions')
    axpoisson_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['chisq']:
    axchisq_mu_n.plot(nn,distributions['chisq'][nn].mean(),'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.set_title('Means for Chi-Square Distributions')
    axchisq_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axgamma_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['gamma']:
    axgamma_mu_n.plot(nn,distributions['gamma'][nn].mean(),'o')
    axgamma_mu_n.set_xlabel('sample size')
    axgamma_mu_n.set_ylabel('sample mean')
    axgamma_mu_n.set_title('Means for Gamma Distributions')
    axgamma_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axexponential_mu_n = fig_mu_N.add_subplot(1,1,1)
for nn in distributions['exponential']:
    axexponential_mu_n.plot(nn,distributions['exponential'][nn].mean(),'o')
    axexponential_mu_n.set_xlabel('sample size')
    axexponential_mu_n.set_ylabel('sample mean')
    axexponential_mu_n.set_title('Means for Exponential Distributions')
    axexponential_mu_n.plot([min(mysizes),max(mysizes)],[mymean,mymean],'k')
```

```python
nbins = np.round(np.sqrt(len(allmeans)))
pl.hist(allmeans,bins = nbins)
n, b, allhist = hist(allmeans,bins = nbins)
bin_max = np.where(n == n.max())
from scipy.stats import norm
param = norm.fit(allmeans)
x = linspace(min(allmeans),max(allmeans),len(allmeans))
y = exp(-((x-param[0])**2)/(2*(param[1]**2)))*n[bin_max][0]
title('Histogram of all Means and Fitted Normal distribution')
xlabel('Means')
ylabel('Frequency')
plot(x,y,'r-')
show()
```
